```python
'''
Multiple Linear Regression
Multiple regression models how a single response(dependent) variable Y depends linearly on a number of
predictor(independent) variables.

Prediction Model Equation: y = c + m1x1 + m2x2 + ..... mnxn
y = response variable (dependent variable)
mn are the coefficients (slopes)
xn are the independent variables
e is error factor, is added to minimize the sum of the errors of prediction
c is the intercept

Formula : to find m and c
m1 = Σx2² * Σx1y - Σx1x2 * Σx2y / Σx1² * Σx2² - (Σx1x2)²
m2 = Σx1² * Σx2y - Σx1x2 * Σx1y / Σx1² * Σx2² - (Σx1x2)²

Example : The selling price of a house can depend on the desirability of the location, the
number of bedrooms, the number of bathrooms, the year the house was built,
the square footage of the lot and a number of other factors.

Ref : http://faculty.cas.usf.edu/mbrannick/regression/Part3/Reg2.html
'''
```

```python
import pandas as pd

df = pd.read_csv("home_prices.csv")
df
```

```python

x1 = list(df['Area'])
x2 = list(df['BHK'])
y = list(df['Price'])

class helper_multiple:
    
    def compute_mean(self,lstnum):
        num_total = 0
        for each in lstnum:
            num_total += each
        return num_total / len(lstnum)
            
    def compute_sum(self,lstnum):
        num_total = 0
        for each in lstnum:
            num_total += each
        return num_total
    

    def compute_sqr(self,x):
        sum_x_sqr = 0
        for each in x:
            sum_x_sqr += each * each
        return sum_x_sqr
    
    
    def compute_sum_x_y(self,x,y):
        sum_xy = 0
        i = 0
        while i < len(x):
            sum_xy += x[i]*y[i]
            i += 1
        return sum_xy
    
    def compute_sum_x_x(self,x1,x2):
        sum_xx = 0
        i = 0
        while i < len(x1):
            sum_xx += x1[i]*x2[i]
            i += 1
        return sum_xx
    
def cal_coeff(x1,x2,y):
    helper = helper_multiple()
    # -------------cal Sum and mean---------------
    sum_X1 = helper.compute_sum(x1)
    sum_X2 = helper.compute_sum(x2)
    sum_Y = helper.compute_sum(y)
    sum_X1_Y = helper.compute_sum_x_y(x1,y)
    sum_X2_Y = helper.compute_sum_x_y(x2,y)
    sum_X1_X2 = helper.compute_sum_x_x(x1,x2)
    y_mean = helper.compute_mean(y)
    x1_mean = helper.compute_mean(x1)
    x2_mean = helper.compute_mean(x2)
    sum_X1_sqr = helper.compute_sqr(x1)
    sum_X2_sqr = helper.compute_sqr(x2)
    #------------substitute sum in equation-----------
    sum_x1_y = sum_X1_Y - ((sum_X1 * sum_Y) / len(y)) # The X and Y are in lower so they are distinct
    sum_x2_y = sum_X2_Y - ((sum_X2 * sum_Y)/len(y))
    sum_x1_x2 = sum_X1_X2 - ((sum_X2 * sum_X1)/len(y))
    sum_x1_sqr = sum_X1_sqr - ((sum_X1 * sum_X1)/len(x1))
    sum_x2_sqr = sum_X2_sqr - ((sum_X2 * sum_X2)/len(x2))
    #------------subsitute these in main formula--------
    b1 =((sum_x2_sqr * sum_x1_y) - (sum_x1_x2 * sum_x2_y))/ ((sum_x1_sqr * sum_x2_sqr) - (sum_x1_x2 * sum_x1_x2))
    b2 = ((sum_x1_sqr * sum_x2_y) - (sum_x1_x2 * sum_x1_y)) /((sum_x1_sqr * sum_x2_sqr) - (sum_x1_x2 * sum_x1_x2))
    c = y_mean-(b1*x1_mean)-(b2*x2_mean)
    print(c,b1,b2)
    return c,b1,b2
        
c, b1, b2 = cal_coeff(x1,x2,y)
var_x1 = int(input("Enter Area : "))
var_x2 = int(input("Enter BHK : "))
#y_res = c + ((var_x1 * b1) + (var_x2 * b2))
#print(y_res)
print("Estimated Price in Lacs : ",round(c + ((var_x1 * b1) + (var_x2 * b2))))
```
